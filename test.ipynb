```python
import heapq
def lazy_dijkstras(graph, root):
    n = len(graph)
    # set up "inf" distances
    dist = [Inf for _ in range(n)]
    # set up root distance
    dist[root] = 0
    # set up visited node list
    visited = [False for _ in range(n)]
    # set up priority queue
    pq = [(0, root)]
    # while there are nodes to process
    while len(pq) > 0:
        # get the root, discard current distance
        _, u = heapq.heappop(pq)
        # if the node is visited, skip
        if visited[u]:
            continue
        # set the node to visited
        visited[u] = True
        # check the distance and node and distance
        for v, l in graph[u]:
            # if the current node's distance + distance to the node we're visiting
            # is less than the distance of the node we're visiting on file
            # replace that distance and push the node we're visiting into the priority queue
            if dist[u] + l < dist[v]:
                dist[v] = dist[u] + l
                heapq.heappush(pq, (dist[v], v))
    return dist
```

```python
from heapq import heappush, heappop

def getAmountOut(amountIn, reserve0, reserve1):
    return 997*amountIn*reserve1 / (1000*reserve0+997*amountIn)

liquidity = {
    ("tokenA", "tokenB"): (17, 10),
    ("tokenA", "tokenC"): (11, 7),
    ("tokenA", "tokenD"): (15, 9),
    ("tokenA", "tokenE"): (21, 5),
    ("tokenB", "tokenC"): (36, 4),
    ("tokenB", "tokenD"): (13, 6),
    ("tokenB", "tokenE"): (25, 3),
    ("tokenC", "tokenD"): (30, 12),
    ("tokenC", "tokenE"): (10, 8),
    ("tokenD", "tokenE"): (60, 25),
}

def find_best_arbitrage_path(start_token, target_token, initial_units):
    visited = set()
    heap = [(0, start_token, initial_units, [])]
    
    while heap:
        profit, token, units, path = heappop(heap)
        
        if token == target_token:
            return profit, path
        
        if token in visited:
            continue
        
        visited.add(token)
        
        for other_token, (reserve0, reserve1) in liquidity.items():
            if token in other_token:
                next_token = other_token[0] if token == other_token[1] else other_token[1]
                next_units = units * reserve1 / reserve0
                next_profit = getAmountOut(units, reserve0, reserve1) - units
                
                if next_token not in visited and next_profit > 0:
                    heappush(heap, (profit + next_profit, next_token, next_units, path + [(token, next_token)]))
    
    return float('-inf'), []

start_token = "tokenB"
target_token = "tokenB"
initial_units = 5

profit, path = find_best_arbitrage_path(start_token, target_token, initial_units)

print("Maximum profit:", profit)
print("Best arbitrage path:", path)
```
